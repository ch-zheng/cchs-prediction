```python
# Import libraries
# Native
import itertools
import os
from pathlib import Path
import shutil
# External
import dlib
import cv2
from PIL import Image
```

```python
# initialize detector, predictor
SHAPE_PREDICTOR_MODEL = 'models/shape_predictor_68_face_landmarks.dat'

detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor(SHAPE_PREDICTOR_MODEL)
```

```python
# variables
src = "E:\\LURIE RESEARCH 08 04 21\\FINAL COHORT 8 17 21\\CCHS"
age_range_list = ["0-1", "2-3", "4-5", "6-8", "9-11", "12-14", "15-17", "18-25", "26-35"]
race_list = ['Caucasian', 'African American', 'Other', 'UNKNOWN']
```

```python
# fix photo orientation
detector = dlib.get_frontal_face_detector()
cnt = 0
for race in race_list:
    for age in age_range_list:
        curr_dir = os.path.join(src, race, age)
        
        for img in os.listdir(curr_dir):
            image = cv2.imread(os.path.join(curr_dir, img))
            faces = detector(image)

            rotated_count = 0
            while rotated_count < 4 and len(faces) == 0:
                image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
                faces = detector(image)
                rotated_count += 1
            if rotated_count < 4:
                cnt += 1
            if rotated_count > 0 and rotated_count < 4:
                cv2.imwrite(os.path.join(curr_dir, img), image)
            print(img)
print(cnt)
```

```python
# which photos not being detected
detector = dlib.get_frontal_face_detector()
cnt = 0
for race in race_list:
    for age in age_range_list:
        curr_dir = os.path.join(src, race, age)
        
        for img in os.listdir(curr_dir):
            image = cv2.imread(os.path.join(curr_dir, img))
            faces = detector(image)

            if len(faces) == 0:
                print(img)
                cnt += 1
print(cnt)
```
